# Feature Engineering

```python
features = copy.copy(df)
```

```python
features['CancellationCode'].value_counts()
```

```python
features['CarrierCancel'] = np.where((features['CancellationCode'] == 'A'), 1, 0)
features['WeatherCancel'] = np.where((features['CancellationCode'] == 'B'), 1, 0)
features['NASCancel'] = np.where((features['CancellationCode'] == 'C'), 1, 0)
features['SecurityCancel'] = np.where((features['CancellationCode'] == 'D'), 1, 0)
```

```python
features['CarrierDelay'] = features['CarrierDelay'].fillna(0)
features['WeatherDelay'] = features['WeatherDelay'].fillna(0)
features['NASDelay'] = features['NASDelay'].fillna(0)
features['SecurityDelay'] = features['SecurityDelay'].fillna(0)
features['LateAircraftDelay'] = features['LateAircraftDelay'].fillna(0)
```

```python
features['TotalDelay'] = features[['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']].sum(axis=1)
```

```python
features['LateFlight'] = np.where((features['TotalDelay'] >= 30), 1, 0)
```

```python
features['LateFlight'].value_counts()
```

```python
features = features.fillna(0)
```

```python
features
```

```python
features.info()
```

```python
features2 = pd.get_dummies(features.drop(['TailNum', 'CancellationCode'], 1))
```

```python
features2.info()
```

## Train-Test Split

```python
X = features2.drop(['TotalDelay', 'LateFlight', 'CarrierDelay',
                     'WeatherDelay', 'NASDelay', 'SecurityDelay',
                     'LateAircraftDelay'], 1)

y = features2['LateFlight']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                    stratify=y, random_state=0)
```

# Models

## Random Forest

```python
rfc = RandomForestClassifier(n_estimators=10, n_jobs=-1)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```

```python
y_pred_rfc = rfc.predict(X_test)
```

```python
cross_rfc = pd.crosstab(y_test, y_pred_rfc)
```

```python
cross_rfc
```

```python
print(classification_report(y_test, y_pred_rfc, digits=4))
```
