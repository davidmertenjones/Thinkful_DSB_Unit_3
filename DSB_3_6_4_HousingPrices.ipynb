```python
df_raw = pd.read_csv('melbourne-housing-market/Melbourne_housing_FULL.csv')

# Too many Ts.

df_raw.rename(columns = {'Lattitude': 'Latitude', 'Longtitude': 'Longitude'}, inplace=True)
```

```python
df_raw.Date = pd.to_datetime(df_raw.Date)
df_raw.Date.head()
```

```python
#Substitute "YearBuilt: 1196" outlier with mean for neighborhood:
df_raw[df_raw['Suburb'] == 'Mount Waverley']['YearBuilt'].mean()
```

```python
df_raw.loc[lambda df: df_raw['YearBuilt'] == 1196, 'YearBuilt'] = 1965
df_raw.loc[lambda df: df_raw['YearBuilt'] == 2019, 'YearBuilt'] = 2018
```

```python
df_raw.describe()
```

```python
df_raw = df_raw.drop(df_raw[df_raw['YearBuilt'] > 2018].index, 0)
```

```python
df_raw.describe()
```

```python
for column in df_raw.columns:
    print(column, (len(df_raw[df_raw[column].isnull()])/len(df_raw)))
```

It looks like we'll lose quite a lot of data if we drop all the NaNs for YearBuilt, so let's try to impute the mean suburb values instead.

```python
df_raw.groupby('Suburb').mean()
```

```python
suburbs = df_raw.groupby('Suburb').mean()['Price'].index
mean_years = np.round(df_raw.groupby('Suburb').mean()['YearBuilt'].values, 0)
```

```python
type(df_raw.groupby('Suburb').mean().loc['Belgrave']['YearBuilt'])
```

```python
mean_years = np.nan_to_num(mean_years)
```

```python
suburbdict = {}

for suburb, year in zip(suburbs, mean_years):
    suburbdict[suburb] = year
```

```python
suburbdict
```

```python
df_raw['YearBuilt'] = df_raw['YearBuilt'].fillna(0)
```

```python
def impute_year(x, y):
    
    """x is year, y is suburb"""
    
    if x > 0:
        year = x
    else:
        year = suburbdict[y]
    return year
```

```python
df_raw['YearImputed'] = [impute_year(x, y) for x, y in zip(df_raw['YearBuilt'], df_raw['Suburb'])]
```

```python
df = df_raw.drop(['YearBuilt', 'Address', 'CouncilArea', 'SellerG'], 1)
```

```python
df = df.dropna()
```

```python
df
```

```python
df.info()
```

```python
for column in df.columns:
    print(column, '\t\t\t',len(df[column].unique()),'\t\t\t', df[column].dtype)
```

```python
features = pd.get_dummies(df)
```

# Model - Price

```python
X = features.drop(['Price', 'Date'], 1)
y = np.log(features['Price'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

## Linear Regression

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
mse = mean_squared_error(y_test, lr.predict(X_test))
```

```python
mse
```

```python
np.sqrt(mse)
```

```python
mae = mean_absolute_error(y_test, lr.predict(X_test))
```

```python
mae
```

```python
y_test.mean()
```
